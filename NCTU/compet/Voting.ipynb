# Voting


```python
import os
```

```python
filess=[36,14,25,27,33,34,39,40,41,38]
#best combination = [36,14,25,27,33,34,39]

path="voting_data/"

labels=[]
#for i in range (last_file+1):
for i in filess:
    if os.path.exists(path+"Result"+str(i)+".csv"):
        f=open(path+"Result"+str(i)+".csv")
        Lines=f.readlines()
        f.close()
        lbl=[]
        Lines=Lines[1:]
        for line in Lines:
            lbl.append(int(line[15]))
        labels.append(lbl)
        
print(len(labels),len(labels[0]))
for i in range(len(labels)):
    print(labels[i][0])

    
```

```python
voting=[]

for i in range (len(labels[0])):
    nb=[0,0,0,0,0,0]
    for j in range (len(labels)):
        nb[labels[j][i]]+=1
    tlbl=[k for k,l in enumerate(nb) if l==max(nb)]
    if len(tlbl)==0:
        tlbl=tlbl[0]
    else:
        if len(tlbl)==len(nb):
            tlbl=labels[0][j]
        else:
            if labels[0][j] in tlbl:
                tlbl=labels[0][j]
            else:
                tlbl=tlbl[0]
    voting.append(tlbl)

print(len(voting))
print(voting)
```

```python
f=open(path+"ResultVoting.csv","w")
f.write("ID,Label\n")
k=0
for line in Lines:
    f.write(line[:15]+str(voting[k])+"\n")
    k+=1
f.close()
    
```
